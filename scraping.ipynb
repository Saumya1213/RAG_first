```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://aryanjalan.com/swiggy-case-study/'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
about = soup.find_all("p")
```

```python
with open("swiggy_info.txt", "a", encoding="utf-8") as text_file:
    for paragraph in about:
      paragraph_text = paragraph.text.strip()
      text_file.write(paragraph_text + "\n\n")
```

```python
# hiring process info

url = 'https://blog.swiggy.com/2022/03/25/what-work-looks-like-at-swiggy/'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')
print(soup)
```

```python
print(about_2)
```

```python
with open("swiggy_info.txt", "a", encoding="utf-8") as text_file:
    for paragraph in about_2:
      paragraph_text = paragraph.text.strip()
      text_file.write(paragraph_text + "\n\n")
```
